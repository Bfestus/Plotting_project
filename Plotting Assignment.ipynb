```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("fuel_econ.csv")

# Display first few rows to inspect data
print(data.head())

numeric_data = data.select_dtypes(include=[float, int])


# Create a correlation matrix for the dataset
corr_matrix = numeric_data.corr()

# Plot the heatmap using seaborn
plt.figure(figsize=(10, 8))  # Set the size of the heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

# Add title
plt.title("Correlation Heatmap of Fuel Economy Data")
# plt.savefig("plot_images/heatmap")
plt.show()

```

### 1. Heatmap interpretation:
The heatmap shows how different features in your dataset (like engine size, fuel efficiency, etc.) are related to each other. The key focus here is how these features affect the GHG (Greenhouse Gas) score.

Positive relationships (red areas): Some features move together. For example, if fuel efficiency has a strong positive relationship with the GHG score, it means that cars with higher fuel efficiency tend to have better (higher) GHG scores, which is good for the environment.

Negative relationships (blue areas): Some features work in opposite directions. For instance, if engine size or weight shows a strong negative relationship with the GHG score, it means bigger or heavier cars tend to have worse (lower) GHG scores, meaning they produce more emissions.

No relationship (neutral colors): Some features don't seem to impact each other much. If a feature has a near-zero relationship with the GHG score, it means it doesn't really help in predicting emissions
### 2. What Columns Would I Drop and Why:
If you're trying to predict the GHG score, some features might not be helpful because they dont strongly influence emissions. These are the columns I would consider dropping:

Fuel Type: If the heatmap shows that the type of fuel (like gas, diesel) doesnt have much impact on the GHG score, you can remove it because it's not adding useful information.

Transmission Type: If the transmission (manual vs automatic) doesnt have a strong link to GHG score, it's better to drop it. It may not play a big role in determining emissions.

Highly similar features: If two features are almost the same, like engine size and number of cylinders, you dont need both. Pick the one that has a stronger impact on the GHG score and drop the other. Keeping both will just complicate the model.

Why drop them: Removing features that don't strongly impact the GHG score makes your model simpler and more focused. It helps the model learn faster and predict better because it's not getting distracted by information that doesnt matter.
